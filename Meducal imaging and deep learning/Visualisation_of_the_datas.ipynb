```python
# I'm going to create a root to the folder containing the images, then i will use sns to look at them

root = 'intracranial_hemorrhagies_dataset/train_jpg/train_jpg'
#Now we have a list containing a root to all the images in the dataset, that's the purpose of the get_image_files function
roots = get_image_files(root)
```
